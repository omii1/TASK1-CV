# IMPLEMENT AN OBJECT DETECTOR WHICH IDENTIFIES THE OBJECT IN AN IMAGE OR VIDEO

```python
import cv2 
import numpy as np
import  matplotlib.pyplot as plt
```

```python
configfile = "ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model = "frozen_inference_graph.pb"
```

```python
model = cv2.dnn_DetectionModel(frozen_model,configfile)
```

```python
classlabels = []
file_name = "lable.txt"
with open(file_name,'r') as f:
    classlabels = f.read().rstrip('\n').split('\n')
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
```

```python
print(classlabels)
```

# FIRST IMAGE :

```python
img = cv2.imread("james bond.jpg")
plt.imshow(img)
```

```python
img = cv2.imread("james bond.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex , confidece , bbox = model.detect(img,confThreshold = 0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 1
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd , conf ,boxes in zip (ClassIndex.flatten(),confidece.flatten(),bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classlabels[ClassInd-1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=1)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
